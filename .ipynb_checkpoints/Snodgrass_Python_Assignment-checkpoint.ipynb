```python
###From Dr. X ###
def get_sequences_from_file(fasta_fn):
    sequence_data_dict = {}
    for record in SeqIO.parse(fasta_fn, "fasta"):
        description = record.description.split()
        species_name = description[1] + " " + description[2]
        sequence_data_dict[species_name] = record.seq
    return(sequence_data_dict)
```

* `sequence_data_dict = {}` Defines a dictionary object named `sequence_data_dict` 
* `for record in SeqIO.parse(fasta_fn, "fasta"):` Creates a for loop utilizing `SeqIO.parse` method
* `fasta_fn` is the file from which these sequences are coming from, "fasta" tells the method to extract the information in fasta format
* `description = record.description.split()` Creates `description` list object from the description line in the fasta file. The `record.description.split()` is used to break up the information found within the single description line by breaking the line into separate strings by spaces. Each resulting index is a string from the description line. 
* `species_name = description[1] + " " + description[2]` The`species_name` becomes a list and is created from `description` using index 1 and index 2 which are the genus and species name (in this case based on where those strings and space are within the original fasta file). The `+ " " +` part of this script ensures there's a space between these strings.
* `sequence_data_dict[species_name] = record.seq
`return(sequence_data_dict)` This is creating an entry for a particular record within the dictionary such that the species_name is the key to the value of record.seq. This iterates via the for loop over every record within the file resulting in a dictionary of every record keyed by genus species names.

Here is my translate function per Dr. X's psuedocode suggestions:

```python
def translate_function(string_nucleotides): #input string_nucleotides must be a Seq object Seq('',IUPAC.unambiguous_dna)
    mito_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"] 
    counter = 0
    aa_seq_string = Seq('',IUPAC.protein)
    for codon in string_nucleotides:
            codon = test_seq[counter:counter+3] #pull out an individual codon
            aa = codon.translate(mito_table) #for a codon compare to the codon table to find out what aa it encodes
            if aa == "*": #if codon encodes a stop codon
                break #stop the loop
            else: #if any other symbol
                aa_seq_string = aa_seq_string + aa #add each aa to a string containing the full aa sequence
                counter = counter + 3 #add 3 to counter to move on to next codon
    return(aa_seq_string) #returns the aa string
```

```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
def compute_molecular_weight(aa_seq):
    aa_seq_as_str = str(aa_seq) #Turns seq object into a string object
    index = 0
    for aa in aa_seq_as_str: #this for loop will cut the string at stop codon if encountered
        if aa == "*":
            aa_seq_as_str = aa_seq_as_str[:index] #if the aa is *, then this slices all the previous AA's stoping at the * and renames original object
        else:
            index += 1 #iterator ensures that the previous AA is included in the final sequence
    analysed_aa_seq = ProteinAnalysis(aa_seq_as_str) #Do the protein analysis package on final sequence
    return(analysed_aa_seq.molecular_weight()) #returns the molecular weight of the sequence
```

### 5  YOUR GC CONTENT ANALYSIS FUNCTION 
#### Write a function that calculates the GC-content (proportion of "G" and "C") of each DNA sequence and returns this value.

```python
def GC_content(string_nucleotides):
    GC_counter = 0 #Start a counter
    for base in string_nucleotides: #For loop to count G's and C's
        if base == "G":
            GC_counter += 1 #If base is a G, increase counter by 1
        if base == "C":
            GC_counter += 1 #If base is a C, increase counter by 1
    GCcontent = GC_counter/len(string_nucleotides) #calculate GC content by # G's and C's counted/total number of bases within string
    return(GCcontent) 
```

Using python we could look at other aspects of the sequences. If doing any comparative genetic or phylogenetic work, sequence alignment will be necessary. However sequence alignment, especially for multiple sequences, can take a lot of memory unless using a command line tool. Here I will attempt to use the Biopython pairwise2 module to try and align _Ursus spelaeus_ and _Ursus maritimus_ (identified as being similar in mass, but what about genetically?) and _Ailuropoda melanoleuca_ and _Ursus thibetanus_ (most distinct from each other in the scatterplot). 

```python
from Bio import pairwise2 #required module (as well as SeqIO that's been previously imported)
from Bio.SubsMat.MatrixInfo import blosum62
Uspelaeus_seq = cytb_seqs['Ursus spelaeus'] #pulling out interesting sequences and saving as seq objects
Umaritimus_seq = cytb_seqs['Ursus maritimus']
Amelanoleuca_seq = cytb_seqs['Ailuropoda melanoleuca']
Uthibetanus_seq = cytb_seqs['Ursus thibetanus']
```

```python
spelaeus2maritimus_alignment = pairwise2.align.globalds(Uspelaeus_seq, #first sequence
                                                        Umaritimus_seq, #second sequence
                                                        blosum62, #matrix used for amino acid sequence match scores
                                                        -10, #open gap penalty
                                                        -0.5) #gap extension penalty
print(pairwise2.format_alignment(*spelaeus2maritimus_alignment[0])) #prints alignment in a visual format
```

The above code for pairwise2 comes from: http://biopython.org/DIST/docs/tutorial/Tutorial.html#sec:pairwise2

The parameters used were given because the source said that this produces better alignments. Additionally if memory limitations are hit with this alignment, there's an option for `score_only`, which when set to `True` will only return the score of the alignment. 

However, it's not clear to me what the score means. So I tried the same alignment but without the specific parameters (only using global defaults). 

```python
test_alignment = pairwise2.align.globalxx(Uspelaeus_seq,Umaritimus_seq)
```
